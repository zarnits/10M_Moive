### (2) 샘플링으로 정규분포인지 확인

```python
np.random.seed(42)
sample_means1 = []
for _ in range(1000):
    m = sample_movie2["상영기간"].sample(30).mean()
    sample_means1.append(m)
```

- 샘플링의 평균값과 sample_movie2["상영기간"]의 평균값 비교

```python
np.mean(sample_means1)
```

```python
sample_movie2["상영기간"].mean()
```

- 2개의 값이 거의 비슷한 것을 확인(신뢰성 up), 정규분포가 아닐 가능성이 크다

### (3) 정규성 검사, shapiro.test

```python
from scipy import stats
from scipy.stats import shapiro
from scipy.stats import probplot
```

```python
shapiro_test = stats.shapiro(sample_movie2["상영기간"])
shapiro_test
```

- ShapiroResult pvalue=0.0073935589753091335
- 0.05 이하이므로 정규분포를 따르지 않는다

```python
shapiro_test = stats.shapiro(sample_movie2["누적관객수"])
shapiro_test
```

- ShapiroResult pvalue=0.0013685855083167553
- 0.05 이하이므로 정규분포를 따르지 않는다

### (4)상영기간과 누적관객수의 상관관계 분석
- 피어슨의 상관계수 : 모수적인 방법
- 스피어만의 상관 계수, 켄달의 tau : 비모수적인 방법


#### 스피어맨의 상관관계분석

```python
from scipy.stats import spearmanr
```

```python
correlation, p_value = spearmanr(sample_movie2['누적관객수'], sample_movie2['상영기간'])
print(f"Spearman correlation coefficient: {correlation}")
print(f"P-value: {p_value}")
```

#### 켄달의 상관관계분석

```python
from scipy.stats import kendalltau
```

```python
corr, p_value = kendalltau(sample_movie2['누적관객수'], sample_movie2['상영기간'])
print("Kendall correlation coefficient: ", corr)
print("p-value: ", p_value)
```

### 상위10개와 하위10개의 상영기간과 누적관객수의 상관관계 분석
- 피어슨의 상관계수 : 모수적인 방법
- 스피어만의 상관 계수, 켄달의 tau : 비모수적인 방법

#### 스피어만의 상관계수

```python
from scipy.stats import spearmanr
```

```python
correlation, p_value = spearmanr(top10['상영기간'], top10['누적관객수'])
print(f"Spearman correlation coefficient: {correlation}")
print(f"P-value: {p_value}")
```

```python
correlation, p_value = spearmanr(down10['상영기간'], down10['누적관객수'])
print(f"Spearman correlation coefficient: {correlation}")
print(f"P-value: {p_value}")
```

#### 켄달의 상관계수

```python
from scipy.stats import kendalltau
```

```python
corr, p_value = kendalltau(top10['상영기간'], top10['누적관객수'])
print("Kendall correlation coefficient: ", corr)
print("p-value: ", p_value)
```

- 켄달의 상관계수에 따르면 상영기간 상위10개의 상영기간과 누적관객수의 상관관계는 무의미한 수준이다

```python
corr, p_value = kendalltau(down10['상영기간'], down10['누적관객수'])
print("Kendall correlation coefficient: ", corr)
print("p-value: ", p_value)
```

- 켄달의 상관계수에 따르면 상영기간 하위10개의 상관관계 역시 무의미하다
