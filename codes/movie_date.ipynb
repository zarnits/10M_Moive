```python
nation1 = pd.DataFrame(pd.value_counts(nation_list))
nation1.columns=["Count"]
nation2 = pd.DataFrame(pd.value_counts(M_list))
nation2.columns=["Count"]
nation3 = pd.DataFrame(pd.value_counts(none_10M_list))
nation3.columns=["Count"]
```

```python
fig = plt.figure(figsize=(16,6)) # 그래프 위치 지정

ax1 = fig.add_subplot(131)       # 위치 지정
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

#define Seaborn color palette to use
colors = sns.color_palette('pastel')

#create pie chart
ax1.pie(nation1["Count"], 
        labels=['한국', '미국', '영국', '일본', '호주', '홍콩', '체코', '독일', '프랑스','캐나다' ], 
        colors = colors, autopct='%.0f%%')
ax2.pie(nation2["Count"], 
        labels=['한국', '미국', '홍콩', '영국'], colors = colors, autopct='%.1f%%')
ax3.pie(nation3["Count"], 
        labels=['한국', '미국', '체코', '독일'], colors = colors, autopct='%.1f%%')

ax1.set_title("상위권 영화(1~200위)")
ax2.set_title("천만 영화(1~28위)")
ax3.set_title("비천만 영화(29~58위)")
```
