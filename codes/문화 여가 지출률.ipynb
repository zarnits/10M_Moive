# 평일/ 휴일 문화예술 관람활동 데이터

```python
weekday = pd.read_csv("C:\\Users\\user\\Downloads\\평일 문화예술 관람활동.csv", encoding="utf-8")
```

```python
holiday = pd.read_csv("C:\\Users\\user\\Downloads\\휴일 문화예술 관람활동.csv", encoding="utf-8")
```

```python
del weekday['통계분류(1)']
```

```python
del holiday['통계분류(1)']
```

```python
weekday.rename(columns={"영화관람":"평일 영화관람률"}, inplace=True)
```

```python
holiday.rename(columns={"영화관람":"공휴일 영화관람률"}, inplace=True)
```

```python
movie_watch = pd.merge(weekday, holiday, how='left')
```

```python
movie_watch
```

```python
movie_watch.rename(columns={"시점":"연도"}, inplace=True)
```

```python
movie_watch['영화관람률'] = (movie_watch['표본수'] * movie_watch['평일 영화관람률'] + movie_watch['표본수'] * movie_watch['공휴일 영화관람률'])/movie_watch['표본수'] 
```

```python
movie_watch['연도'] = movie_watch['연도'].astype(str)
```

# 문화여가지출률과 평일/공휴일 관람활동 merge

```python
final = pd.merge(leisure_consume, movie_watch, how='outer')
```

```python
final.fillna(0)
```

```python
final.to_csv("leisure_expenditure.csv", encoding="utf-8", index=False)
```
