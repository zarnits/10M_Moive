## 가장 앞에있는 1987로 테스트, 모양 확인

```python
a = pd.read_csv(f"C:/GitTest/10M_Movie/movie_DB/individual_trend/{m_list[0]}")
a
```

- iloc로 숫자 형식 접근 1번째 열의 주소인 0과 
- 맨 처음 행의 주소 (개봉일) 0으로 접근 
- 마지막 행의 주소 (스크린 점유율 1% 미만인 임의의 상영종료일) -1로 접근 

```python
print(a.iloc[0,0])
print(a.iloc[-1,0])
```

- 모든 csv 파일이 있는 영화들의 개봉일, 상영 종료일을 구하는 반복문

```python
start_date = []
end_date = []

for i in m_list:
        
    # 파일 읽어오기
    movie = pd.read_csv(f"C:/GitTest/10M_Movie/movie_DB/individual_trend/{i}")
    start_date.append(movie.iloc[0,0])
    end_date.append(movie.iloc[-1,0])
        
```

## 상영시작일과 종료일을 모은 데이터 프레임 생성

- 영화 파일들의 이름에서 .csv제거해서 이름 추출하는 반복문

```python
a_list = []
for i in m_list:
    a_list.append(i[0:-4])

a_list
```

### 영화, 상영시작일, 상영종료일 가지는 데이터 프레임 생성

```python
date_df = pd.DataFrame({'영화':a_list, '상영시작일':start_date, '상영종료일':end_date})
```

```python
date_df.to_csv("C:/GitTest/10M_Movie/movie_DB/individual_movie/movie_date.csv", index=False)
```
